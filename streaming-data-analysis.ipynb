# Identificação e correção de problemas gerais

## Tratamento de dados nulos

```python
# Verificando valores nulos
print("Valores nulos - Netflix")
print(df_netflix.isnull().sum())
print("\n")
print("Valores nulos - Disney+")
print(df_disney.isnull().sum())
print("\n")
print("Valores nulos - Amazon Prime")
print(df_amazon.isnull().sum())
print("\n")
print("Valores nulos - Hulu")
print(df_hulu.isnull().sum())
print("\n")
```

```python
# Tratamento de dados nulos - Netflix
df_netflix[["director", "cast", "country"]] = df_netflix[["director", "cast", "country"]].fillna("anonymous")
df_netflix['duration'] = df_netflix['duration'].fillna('unknown')
df_netflix['rating'] = df_netflix['rating'].fillna('UNRATED')
```

```python
# Tratamento de dados nulos - Disney+
df_disney[["director", "cast", "country"]] = df_disney[["director", "cast", "country"]].fillna("anonymous")
df_disney['rating'] = df_disney['rating'].fillna('NR')
```

```python
# Tratamento de dados nulos - Amazon Prime
df_amazon[["director", "cast", "country"]] = df_amazon[["director", "cast", "country"]].fillna("anonymous")
df_amazon['rating'] = df_amazon['rating'].fillna('UNRATED')
```

```python
# Tratamento de dados nulos - Hulu
df_hulu[["director", "cast", "country"]] = df_hulu[["director", "cast", "country"]].fillna("anonymous")
df_hulu['duration'] = df_hulu['duration'].fillna('unknown')
df_hulu['rating'] = df_hulu['rating'].fillna('UNRATED')
df_hulu['description'] = df_hulu['description'].fillna('unknown')
```

**Tratamento dos valores nulos da coluna "data_added"**

Para essa coluna pode-se adotar diferentes abordagens

*Preencimento com uma data aproximada*
- Se o título tem um ano de lançamento (release_year) conhecido, pode-se assumir que ele foi adicionado no mesmo ano. Entretando esse abordagem assume que todos os títulos foram adicionados ao catálogo no mesmo ano de lançamento, o que não é sempre verdadeiro.

*Criar uma categoria "Desconhecido" (unknown)*
- Preencher essa coluna como desconhecido mantém os dados consistentes porque evita a introdução de vieses ao adicionar uma data aproximada, todavia não permite análises baseadas na data.

*Exlcuir registros com dados nulos*
- Caso os valores nulos na coluna fossem pequenos essa seria uma boa opção, mas ela não é ideal para o dataset da Amazon Prime, haja vista que tem 9513 sem registros.


**O que será feito**

*Preenchimento com "unknown"*
- Preencher todos os valores com "unknown" é a abordagem mais consistente nesse caso, porque evita o viés e mantém a integridade dos dados para análises futuras.

```python
# Netflix
df_netflix["date_added"] = df_netflix["date_added"].fillna("unknown")

# Disney+
df_disney["date_added"] = df_disney["date_added"].fillna("unknown")

# Amazon Prime
df_amazon["date_added"] = df_amazon["date_added"].fillna("unknown")

# Hulu
df_hulu["date_added"] = df_hulu["date_added"].fillna("unknown")
```

# Identificação e correção de problemas particulares

## Netflix

```python
# Visualizando as informações que estão contidas nas colunas
for column in df_netflix.columns:
    print(f"Coluna: {column}")
    print(df_netflix[column].unique())
    print("-"*50)
```

```python
# Verificando a padronização dos ids na coluna "show_id"
pattern = r"^s\d+$"
invalid_rows = df_netflix[~df_netflix["show_id"].str.match(pattern, na=False)]
print("Registros que não seguem o padrão 'sXXX'")
invalid_rows
```

```python
# Verificando se há valores fora do esperado na coluna "type"
expected_values = ["Movie", "TV Show"]
invalid_type_rows = df_netflix[~df_netflix["type"].isin(expected_values)]
print("Registros com valores inesperados na coluna 'type':")
invalid_type_rows
```

```python
# Verificando as linhas que estão com valores incorretos na coluna "rating"
duration_pattern = r"^\d+\smin$"
invalid_rating_rows = df_netflix[df_netflix["rating"].str.match(duration_pattern, na=False)]
print("Linhas com valores inconsistentes na coluna 'rating':")
invalid_rating_rows
```

```python
# Movendo os valores de "rating" para a coluna "duration"
invalid_rows = df_netflix["rating"].str.match(duration_pattern, na=False)
df_netflix.loc[invalid_rows, "duration"] = df_netflix.loc[invalid_rows, "rating"]
df_netflix.loc[invalid_rows, "rating"] = 'UNRATED' 
print("Verificando a correção:")
print(df_netflix["rating"].unique())
```

**Coluna "Rating"**
